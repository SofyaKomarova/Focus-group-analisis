```python
fname = input('Enter File: ')
if len(fname) < 1 : fname = 'Focus_Group_Transcribtion.txt'
hand = open(fname)


wordlis = []
indexlis=[]
di = dict()

for lin in hand:
    wordlis.append(lin)
    lin = lin.rstrip()
    wds = lin.split()
    for w in wds:
        di[w] = di.get(w,0)+1
        

#print(wordlis)

def occurences(word,wordlis,n1,n2):
    for i in range(len(wordlis)):
        if wordlis[i]==word:
            indexlis.append(i)
    for elem in indexlis:
        print(wordlis[elem-n1:elem+n2+1])
```

```python
occurences('immagine',wordlis,3,4)
```
